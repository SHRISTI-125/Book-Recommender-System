```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_ratings_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns = {'Book-Rating' : 'num_ratings'}, inplace = True)
num_ratings_df
```

```python
avg_ratings_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings_df.rename(columns = {'Book-Rating' : 'avg_ratings'}, inplace = True)
avg_ratings_df
```

```python
popular_df = num_ratings_df.merge(avg_ratings_df, on = 'Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending = False)
```

```python
popular_df.head(20)
```

```python
popular_df.merge(books, on = 'Book-Title')
```

```python
popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_ratings']]
```

## Collaborative Based Filtering

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
book_readers = x[x].index
```

```python
filtering_rating = ratings_with_name[ratings_with_name['User-ID'].isin(book_readers)]
```

```python
filtering_rating
```

```python
y = filtering_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_book = y[y].index
```

```python
famous_book
```

```python
final_ratings = filtering_rating[filtering_rating['Book-Title'].isin(famous_book)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0, inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt).shape
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    # distance = similarity_score[index]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x : x[1], reverse = True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    # distance = similarity_score[index]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x : x[1], reverse = True)[1:6]
    
    data = []
    
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('The Hours : A Novel')
```
